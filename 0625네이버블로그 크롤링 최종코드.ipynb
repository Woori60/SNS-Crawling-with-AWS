```python
from bs4 import BeautifulSoup  
import requests                
import urllib.request as req        #해당 3가지 library는 알 것이라 생각하여 설명 생략

from itertools import count         # 반복횟수가 불확실한 for Loop에서 사용
from collections import OrderedDict # 마지막 게시글인지 여부를 비교확인하기 위해 사용 
import re                           # 정규 표현식을 사용하기 위해 사용[출처] [파이썬]블로그 게시글 크롤링(crawling)_01(개발 환경)[BeautifulSoup,Requests]|작성자 GUN


url = 'http://search.naver.com/search.naver'
hrd = {'User-Agent': 'Mozilla/5.0' , 'referer' : 'http://naver.com'}

param = {'where' : 'post',
         'query' : '여행',
        'date_from' : '20170625',
        'date_to' : '20190625'}

response = requests.get(url,params=param,headers=hrd)

soup = BeautifulSoup(response.text, 'html.parser')

area = soup.find('div', {'class': 'blog section _blogBase _prs_blg'}).find_all('a', {'class': 'url'})

#for tag in area:
#    url1=tag.get('href')
#    print("{:}".format(url1))
    
    
f = open('link.txt', 'w')    
for tag in area:
    url1=tag.get('href')
    f.write("{:}".format(url1))
    f.write('\n')
f.close()

#다음, 티스토리 제외하기
mf = open('modlink.txt', 'w')


def get_final_url(tmp):
    try:
        if (tmp.startswith("https://blog.naver.com") or tmp.startswith("http://blog.naver.com")):
            html_result = requests.get(tmp)
            soup_temp = BeautifulSoup(html_result.text, 'html.parser')
            area_temp = soup_temp.find(id='mainFrame')
            url_3 = area_temp.get('src')
            return "https://blog.naver.com" + url_3

        else:
            try:
                html_result = requests.get(tmp)
                soup_temp = BeautifulSoup(html_result.text, 'html.parser')
                area_temp = soup_temp.find(id='screenFrame')
                url_2 = area_temp.get('src')

            except:
                try:
                    area_temp = soup_temp.find(id='mainFrame')
                    url_3 = area_temp.get('src')
            
                    url_4 = "https://blog.naver.com" + url_3
                except:
                    return None
            try:
                html_result = requests.get(url_2)
                soup_temp = BeautifulSoup(html_result.text, 'html.parser')
                area_temp = soup_temp.find(id='mainFrame')
                url_3 = area_temp.get('src')
                
                url_4 = "https://blog.naver.com" +url_3
                return url_4
            except :
                print("error")
                return None
        
    except :
        print("error")
        return None
    

for tag in area:
    url1=tag.get('href')
    mf.write("{:}".format( get_final_url(url1)))
    mf.write('\n')

mf.close()
```
